```python
import pandas as pd

df = pd.read_csv('contestants.csv', encoding='utf-8')
```

```python
# cf. the discussion here: https://www.reddit.com/r/MachineLearning/comments/bf137p/p_python_package_to_easily_retrain_openais_gpt2/
# gpt2.generate(sess, prefix="<|startoftext|>", truncate="<|endoftext|>")

def preprocess_lyrics(lyrics, prefix = "<|startoftext|>", suffix = "<|endoftext|>"): 
    lyrics = lyrics.replace('\\n', '. ')
    return f"{prefix} {lyrics} {suffix}"

preprocessed_lyrics = df['lyrics'].apply(preprocess_lyrics)
```

```python
training_data = "\n".join(preprocessed_lyrics)
print(len(training_data))
```

```python
with open('training_data.txt', 'wt', encoding='utf-8') as f:
    f.write(training_data)
```
